### **Central Limit Theorem**

**Scenario**

The goal is to estimate the workload for the management of returns, with the assumption that the number of times for parcel that is received from stores follows a normal distribution.

```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.graphics.gofplots as sm
```

```python
np.random.seed(0)
```

**Normal Distribution**

```python
# Parameters 

mu = 23
sigma = 7
plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.title('Distribuzione Normale : N(23,7)')
plt.show()
```

**Probability n>30 items** 

```python
# Parameters
mu = 23
sigma = 7

# Set up of probability function
stats.norm(mu, sigma)

# Probability X = 30
x1 = 30

# Cumulative Density Function
y1 = stats.norm(mu, sigma).cdf(x1)

# Probability of Density Function
z1 = stats.norm(mu, sigma).pdf(x1)

plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.axvline(x1, color='red')
plt.axhline(z1, color='red')
plt.title('Distribuzione Normale : N(23,7)')
print("Probabilità di avere {} items/cartone è {}%".format(x1, round(100*y1,2)))
plt.show()
```

**Probability = 95%**

```python
# Parameters
mu = 23
sigma = 7

# Set up of probability function
stats.norm(mu, sigma)

# Probability X = 30
y1 = 0.95

# Percent point function (reverse of cumulative density function)
x1 = stats.norm(mu, sigma).ppf(y1)

# Probability of density function 
z1 = stats.norm(mu, sigma).pdf(x1)

plt.figure(figsize=(12, 6))
ax = plt.gca()
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.axvline(x1, color='red')
plt.axhline(z1, color='red')
plt.title('Distribuzione Normale : N(23,7)')
print("Si ha il 95% di probabilità che X <= {} items/cartone".format(round(x1,3)))
plt.show()
```
